```python
char_pairs = {"(": ")", "[": "]", "{": "}", "<": ">"}

incomplete_lines = []
for sequence in inputs:
    open_chars = []
    bad = False
    for char in sequence:
        if char in char_pairs.keys():
            open_chars.append(char)
        else:
            open_char = open_chars.pop()
            if char != char_pairs[open_char]:
                bad = True
    
    if not bad:
        incomplete_lines.append(sequence)
```

```python
end_score_map = {")": 1, "]": 2, "}": 3, ">": 4}

scores = []
for line in incomplete_lines:
    open_chars = []
    for char in line:
        if char in char_pairs.keys():
            open_chars.append(char)
        else:
            open_chars.pop()
    
    closing_chars = []
    for open_char in reversed(open_chars):
        closing_chars.append(char_pairs[open_char])

    score = 0
    for char in closing_chars:
        score *= 5
        score += end_score_map[char]
    
    scores.append(score)
```
