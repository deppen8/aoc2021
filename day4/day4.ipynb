```python
import numpy as np
```

```python

class Board:
    def __init__(self, arr: np.ndarray) -> None:
        self.board: np.ndarray = arr
        self.marked: np.ndarray = np.zeros_like(self.board)
        self.win_idxs = []
        self.won = False
    
    def __repr__(self) -> str:
        return self.board.__repr__()

    def mark(self, num_called: int) -> None:
        self.marked[self.board == num_called] = 1

    def check(self) -> None:
        for axis in [0, 1]:
            sums = self.marked.sum(axis=axis)
            winning_pos = np.where(sums == 5)[0]
            if winning_pos.shape[0] > 0:
                print("WINNING BOARD:")
                print(self.board)
                print("----------")
        
                if axis==0:
                    print("winning column:")
                    print(f"{axis = }")
                    print(f"{winning_pos = }")
                    print(f"{self.board[:, winning_pos[0]]}")
                else:
                    print("winning row:")
                    print(f"{self.board[winning_pos[0], :]}")
                
                return True
```

```python
draws = [13,47,64,52,60,69,80,85,57,1,2,6,30,81,86,40,27,26,97,77,70,92,43,94,8,78,3,88,93,17,55,49,32,59,51,28,33,41,83,67,11,91,53,36,96,7,34,79,98,72,39,56,31,75,82,62,99,66,29,58,9,50,54,12,45,68,4,46,38,21,24,18,44,48,16,61,19,0,90,35,65,37,73,20,22,89,42,23,15,87,74,10,71,25,14,76,84,5,63,95]
```

```python
with open("./data.txt", 'r') as infile:
    lines = infile.readlines()
```

```python
boards = []
board = []
for line in lines:
    if line=="\n":
        boards.append(np.array(board))
        board = []
    else:
        board.append([int(val) for val in line.rstrip().split()])

boards.append(np.array(board)) 
```

```python
# assert all(isinstance(b, Board) for b in boards)
```

```python
def play(draws, boards):
    board_objs = [Board(board) for board in boards]
    for idx, draw in enumerate(draws):
        for board in board_objs:
            if not board.won:
                board.mark(draw)
                if board.check():
                    board.won = True
                    board.win_idxs.append(idx)
                    print(f"{draw = } ({idx}th draw)")
                    print("Game won!")

    return board_objs


winning_boards = play(draws, boards)
```

```python
last_winner_idx = 0
last_winner = None
for board in winning_boards:
    if board.win_idxs[0] > last_winner_idx:
        last_winner_idx = board.win_idxs[0]
        last_winner = board

print(last_winner_idx)
last_winner
```
